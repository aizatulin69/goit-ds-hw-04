```
import random
import numpy as np

#Завдання 1
def task1():
    return f"1: \n {[i for i in range(1, 10)]}\n"

#Завдання 2
def task2():
    return f"2: \n {[[0]*3 for _ in range(3)]}\n"

#Завдання 3
def task3():
    return f"3: \n {[random.randint(1, 10) for _ in range(5)]}\n"

#Завдання 4
def task4():
    return f"4: \n {[random.uniform(0, 1) for _ in range(4)]}\n"

#Завдання 5
def task5():
    a, b = ([random.randint(1, 10) for _ in range(5)],
            [random.randint(1, 10) for _ in range(5)])
    add = [a[i] + b[i] for i in range(5)]
    sub = [a[i] - b[i] for i in range(5)]
    mul = [a[i] * b[i] for i in range(5)]
    return f"5: \n a+b={add} \n a-b={sub} \n a*b={mul}\n"

#Завдання 6
def task6():
    a, b = ([random.randint(0, 20) for _ in range(7)],
            [random.randint(0, 20) for _ in range(7)])
    return f"5: \n {sum(x * y for x, y in zip(a, b))}\n"

# Завдання 7
def task7():
    A = [[random.randint(1,10) for _ in range(3)] for _ in range(2)]
    B = [[random.randint(1,10) for _ in range(2)] for _ in range(3)]
    C = np.dot(A,B).tolist()
    return f"7: \n A={A}\n B={B}\n A*B={C}\n"

# Завдання 8
def task8():
    M = [[random.randint(1,10) for _ in range(3)] for _ in range(3)]
    invM = np.linalg.inv(M).tolist()
    return f"8: \n M={M}\n M^-1={invM}\n"

# Завдання 9
def task9():
    M = [[random.random() for _ in range(4)] for _ in range(4)]
    T = np.transpose(M).tolist()
    return f"9: \n M={M} \n M^T={T} \n"

# Завдання 10
def task10():
    M = [[random.randint(1,10) for _ in range(4)] for _ in range(3)]
    v = [random.randint(1,10) for _ in range(4)]
    result = np.dot(M,v).tolist()
    return f"10: \n M={M} \n v={v} \n M*v={result} \n"

# Завдання 11
def task11():
    M = [[random.random() for _ in range(3)] for _ in range(2)]
    v = [random.random() for _ in range(3)]
    result = np.dot(M,v).tolist()
    return f"11: \n M={M} \n v={v} \n M*v={result} \n"

# Завдання 12
def task12():
    A = [[random.randint(1,10) for _ in range(2)] for _ in range(2)]
    B = [[random.randint(1,10) for _ in range(2)] for _ in range(2)]
    elem_mul = [[A[i][j]*B[i][j] for j in range(2)] for i in range(2)]
    return f"12: \n A={A} \n B={B} \n A.*B={elem_mul} \n"

# Завдання 13
def task13():
    A = [[random.randint(1,10) for _ in range(2)] for _ in range(2)]
    B = [[random.randint(1,10) for _ in range(2)] for _ in range(2)]
    prod = np.dot(A,B).tolist()
    return f"13: \n A={A} \n B={B} \n A*B={prod} \n"

# Завдання 14
def task14():
    M = [[random.randint(1,100) for _ in range(5)] for _ in range(5)]
    total = sum(sum(row) for row in M)
    return f"14: \n M={M} \n Sum={total} \n"

# Завдання 15
def task15():
    A = [[random.randint(1,10) for _ in range(4)] for _ in range(4)]
    B = [[random.randint(1,10) for _ in range(4)] for _ in range(4)]
    diff = [[A[i][j]-B[i][j] for j in range(4)] for i in range(4)]
    return f"15: \n A={A} \n B={B} \n A-B={diff} \n"

# Завдання 16
def task16():
    M = [[random.random() for _ in range(3)] for _ in range(3)]
    row_sums = [[sum(row)] for row in M]
    return f"16: \n M={M} \n Row sums={row_sums} \n"

# Завдання 17
def task17():
    M = [[random.randint(0,10) for _ in range(4)] for _ in range(3)]
    M_sq = [[x**2 for x in row] for row in M]
    return f"17: \n M={M} \n M squared={M_sq} \n"

# Завдання 18
def task18():
    v = [random.randint(1,50) for _ in range(4)]
    v_sqrt = [x**0.5 for x in v]
    return f"18: \n v={v} \n Sqrt={v_sqrt}"

print(task1())
print(task2())
print(task3())
print(task4())
print(task5())
print(task7())
print(task8())
print(task9())
print(task10())
print(task11())
print(task12())
print(task13())
print(task14())
print(task15())
print(task16())
print(task17())
print(task18())

```
